```python
def plotly_line_modes(x,y,number,legend_name='',title_name='',xaxis_name='',yaxis_name=''):#
    if (number==1):
        fig=go.Figure(data=go.Scatter(x=x,y=y,mode='markers+lines',name=legend_name,line=dict(color='royalblue', width=2)))
        fig.update_layout(title=title_name,xaxis_title=xaxis_name,yaxis_title=yaxis_name)
    else:
        fig=go.Figure()
        for i in range(number):
            fig.add_trace(go.Scatter(x=x,y=y.iloc[:,i],name=y.iloc[:,i].name))
        fig.update_layout(title=title_name,xaxis_title=xaxis_name,yaxis_title=yaxis_name)
        fig.show()
    return fig
```

```python
def label_lines_annotations (x_data,y_data,number,labels,colors=['rgb(67,67,67)', 'rgb(115,115,115)', 'rgb(49,130,189)', 'rgb(189,189,189)'],mode_size = [8, 8, 12, 8],line_size=[2, 2, 4, 2],title='',source_text=''):#
    fig=go.Figure()#pandasDataFrame
    for i in range(number):
        
        fig.add_trace(go.Scatter(x=x_data, y=y_data.iloc[:,i], mode='lines',
            name=labels[i],
            line=dict(color=colors[i], width=line_size[i]),
            connectgaps=True,
        ))

        #end points 
        fig.add_trace(go.Scatter(x=[x_data[0],x_data[-1]],
                                 y=[y_data.iloc[:,i][0],y_data.iloc[:,i][-1]],
                                 mode='markers',
                                 marker=dict(color=colors[i],size=mode_size[i])
                                ))
    fig.update_layout(xaxis=dict(showline=True,
                                showgrid=True,
                                showticklabels=True,
                                linecolor='rgb(204,204,204)',
                                linewidth=2,
                                ticks='outside',
                                tickfont=dict(
                                family='Arial',
                                size=12,
                                color='rgb(82, 82, 82)',
                                ),
                        ),
                    yaxis=dict(showgrid=True,zeroline=True,showline=True,showticklabels=False,),
                    autosize=False,
                    margin=dict(autoexpand=True,l=100,r=20,t=110,),
                    showlegend=False,
                    plot_bgcolor='white')
    annotations=[]
    
    #adding labels 
    for y_trace, label, color in zip(np.array(y_data), labels, colors):
        # labeling the left_side of the plot
        annotations.append(dict(xref='paper', x=0.05, y=y_trace[0],
                                  xanchor='right', yanchor='middle',
                                  text=label + ' {}%'.format(y_trace[0]),
                                  font=dict(family='Arial',
                                            size=16),
                                  showarrow=False))
        # labeling the right_side of the plot
        annotations.append(dict(xref='paper', x=0.95, y=y_trace[-1],
                                  xanchor='left', yanchor='middle',
                                  text='{}%'.format(y_trace[-1]),
                                  font=dict(family='Arial',
                                            size=16),
                                  showarrow=False))
        #title
    annotations.append(dict(xref='paper', yref='paper', x=0.5, y=-0.1,
                              xanchor='center', yanchor='top',
                              text=source_text,
                              font=dict(family='Arial',
                                        size=12,
                                        color='rgb(150,150,150)'),
                              showarrow=False))
    fig.update_layout(annotations=annotations)
    return fig
#pandas
    
```
